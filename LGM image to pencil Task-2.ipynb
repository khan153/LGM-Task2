```python
#importing required library 
import cv2 #NOTE-if it is cannot work go to anaconda promp type "python -m pip install opencv-python "
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#to check the size of image for resize if required
image= Image.open('download.jpeg')
print(image.size)
```

```python
#use to read the image using opencv
img=cv2.imread('download.jpeg')
```

```python
#Displaying image using OpenCV is not very straight forward. The following command can be used to display the photo.
im_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(im_rgb)
```

```python
#Convert BGR to Gray: Method 1
#Using cvtColor method of OpenCV.
grayed=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(grayed,cmap='gray')
                    
```

```python
#use for inverted image
inverted=255-grayed
plt.imshow(inverted,cmap='gray')
```

```python
#use for blur image using GaussianBlur Method
blurred=cv2.GaussianBlur(inverted,(21,21),0)
plt.imshow(blurred,cmap='gray')
```

```python
sketch=cv2.divide(grayed,255-blurred,scale=255)
plt.imshow(sketch,cmap='gray')
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(im_rgb)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
